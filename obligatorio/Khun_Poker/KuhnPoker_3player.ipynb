```python
from games.kuhn_poker.kuhn3 import KuhnPoker3
from agents.counterfactualregret_t import CounterFactualRegret
from collections import OrderedDict
from torneo_khun_poker_3 import (
    KuhnPokerAnalyzer,
    analyze_with_agent_dict
)
```

```python
g = KuhnPoker3()
```

```python
print(f"Agentes en el juego: {g.agents}")
print(f"Cantidad: {len(g.agents)}")
```

## Entrenamiento corto (10 iteraciones)

```python
agent_classes = [ CounterFactualRegret, CounterFactualRegret,CounterFactualRegret  ]
my_agents = {}
g.reset()
for i, agent in enumerate(g.agents):
    my_agents[agent] = agent_classes[i](game=g, agent=agent)
```

```python
for agent in g.agents:
    print('Training agent ' + agent)
    my_agents[agent].train(10)
    print('Agent ' + agent + ' policies:')
    print(OrderedDict(map(lambda n: (n, my_agents[agent].node_dict[n].policy()), sorted(my_agents[agent].node_dict.keys()))))
    print('')
```

### Partida corta (1.000 partidas)

```python
analyzer = analyze_with_agent_dict(my_agents, '3p', 1000)
```

### Partida larga (10.000 partidas)

```python
analyzer = analyze_with_agent_dict(my_agents, '3p', 10000)
```
